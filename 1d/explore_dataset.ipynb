```python
# USE WITH CAUTION AND PROBABLY ONLY ONE IN A LIFETIME
# Add duration and sample rate information to the metadata
metadata['duration'] = metadata['end'] - metadata['start']
metadata['sample_rate'] = 0
import os
import librosa
for fold_num in range(1,11):
    fold_dir = dataset_dir + 'audio/fold' + str(fold_num)
    print(fold_dir)
    for audiofile in metadata.slice_file_name[metadata.fold == fold_num]:
        print(audiofile)
        _, sr = librosa.load('../../data/UrbanSound8K/audio/fold'+str(fold_num)+'/'+audiofile, sr=None)
        metadata.loc[metadata.slice_file_name==audiofile, 'sample_rate'] = sr
metadata.to_csv(dataset_dir + '/metadata/UrbanSound8K_added-dur_sr.csv')
```
